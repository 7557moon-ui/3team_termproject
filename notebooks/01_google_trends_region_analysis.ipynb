# 구글 트렌드 지역별 가전제품 검색량 정리

```python
# 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 원본 데이터 로드
df_trends = pd.read_csv("../data/processed/google_trends_region_2021_2025.csv", encoding='utf-8')

# 영어 행정구역 컬럼명을 한글로 매칭
df_trends = df_trends.rename(columns={'geoName': '지역'})

df_trends
```

```python
# 예상 인사이트: 여름철 필수 냉방 가전인 에어컨과 선풍기는 전국 시도별로 관심도의 평균 분포가 다르게 나타날 것이다.
# 가설 검증을 위해 2025년 최신 연도의 가전제품 검색량 컬럼만 리스트로 묶어 별도 분석을 진행한다.

aircon_2025_mean = df_trends.groupby("지역")["에어컨_2025"].mean().reset_index(name="에어컨 평균 검색량")
aircon_2025_mean
# 각 행정구역별 기후 특성에 따른 에어컨 관심도 기본 정렬 자료 생성 완료.
```

```python
# 한글 폰트 깨짐 방지 및 마이너스 기호 설정 (Windows 환경 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False


# 지역 이름이 한눈에 보이도록 평균 검색량이 높은 순서대로 정렬
aircon_plot_data = aircon_2025_mean.sort_values(by="에어컨 평균 검색량", ascending=True)

plt.figure(figsize=(10, 8))
plt.barh(aircon_plot_data["지역"], aircon_plot_data["에어컨 평균 검색량"], color='salmon', edgecolor='black', height=0.6)

plt.title("2025년 대한민국 16개 시도별 에어컨 평균 검색 관심도 비교", fontsize=14, pad=15)
plt.xlabel("평균 검색량 지표 (Relative Search Volume)", fontsize=12)
plt.ylabel("행정구역 (지역)", fontsize=12)
plt.grid(axis='x', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# 각 행정구역별 기후적 특성(내륙 분지 지형 여부, 도심 열섬 현상 등)에 따라 에어컨 가전에 대한 소비자들의 관심도가 상이하게 분포함을 알 수 있다.
# 이 지리적 관심도 격차 자료는 지역별 계절가전 물류 배치 및 타겟 마케팅 가중치를 설정하는 정량적 근거로 활용 가능하다.
```

```python
years = ['2021', '2022', '2023', '2024', '2025']

aircon_years = [df_trends[f"에어컨_{year}"].mean() for year in years]
fan_years = [df_trends[f"선풍기_{year}"].mean() for year in years]
blanket_years = [df_trends[f"전기장판_{year}"].mean() for year in years]

# 3. 냉난방 가전 연도별 트렌드 연계 시각화
plt.figure(figsize=(12, 6))

plt.plot(years, aircon_years, marker='o', color='crimson', linewidth=2.5, label='에어컨 (냉방)')
plt.plot(years, fan_years, marker='s', color='orange', linewidth=2.5, label='선풍기 (냉방)')
plt.plot(years, blanket_years, marker='X', color='dodgerblue', linewidth=2.5, label='전기장판 (난방)')

plt.title("주요 계절가전(냉방 vs 난방) 연도별 전국 평균 검색 트렌드 추이 (2021~2025)", fontsize=14, pad=15)
plt.xlabel("연도", fontsize=12)
plt.ylabel("전국 평균 검색 관심도 지표", fontsize=12)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# 에어컨, 선풍기, 전기장판의 5개년 전국 검색 추이를 연계하여 비교한 결과, 특정 연도의 기후 변동(예: 역대급 폭염이나 이른 한파 등)에 따라 냉난방 가전의 사회적 관심도가 어떻게 상호작용하는지 확인할 수 있다.
# 선풍기처럼 진입장벽이 낮은 필수 가전은 매년 일정한 검색 패턴을 유지하는 반면, 에어컨과 전기장판은 연도별 기후 조건에 따라 관심도가 급격히 교차하거나 동반 상승하는 변동성을 보여준다.
```

```python
df_trends = pd.read_csv("../data/processed/google_trends_region_2021_2025.csv", encoding='utf-8')

# 2. 2025년 최신 기준 가전제품별 지역 데이터 추출
# 전기장판 검색량 기준으로 내림차순 정렬하여 지역별 기후 특성이 한눈에 보이도록 구성
df_2025 = df_trends[['geoName', '에어컨_2025', '선풍기_2025', '전기장판_2025']].sort_values(by='전기장판_2025', ascending=False)
df_2025 = df_2025.set_index('geoName')

# 3. 지역별 주요 계절가전 관심도 비교 (그룹화 막대 그래프)
df_2025.plot(kind='bar', figsize=(15, 7), color=['crimson', 'orange', 'dodgerblue'], edgecolor='black', width=0.8)

plt.title("2025년 전국 16개 시도별 계절가전(냉방 vs 난방) 검색 관심도 비교", fontsize=14, pad=15)
plt.xlabel("행정구역 (지역)", fontsize=12)
plt.ylabel("검색 관심도 지표", fontsize=12)
plt.xticks(rotation=45)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.legend(['에어컨', '선풍기', '전기장판'], fontsize=11)
plt.tight_layout()
plt.show()

# 구글 트렌드의 지역(Region) 특성을 살려 시도별 검색량을 비교한 결과, 전국 평균치 선 그래프에서는 보이지 않던 지리적 소비 편차가 명확히 드러난다.
# 상대적으로 겨울철 추위가 강한 강원도나 일부 내륙 지역에서는 난방기구(전기장판)의 관심도 상자가 확연히 높게 나타나는 반면, 대도시 권역이나 남부 내륙 지역에서는 에어컨과 선풍기 같은 냉방 가전의 강세가 뚜렷하다.
# 이는 해당 데이터셋이 가진 지리적 정보라는 고유 가치를 가장 올바르게 살려낸 시각화 분석이다.
```

```python
df_2025_matrix = df_trends[['geoName', '에어컨_2025', '선풍기_2025', '전기장판_2025']].set_index('geoName')

# 그래프 표기용 컬럼명 매칭
df_2025_matrix.columns = ['에어컨', '선풍기', '전기장판']

# 3. 히트맵 시각화
plt.figure(figsize=(10, 9))

# annot=True로 격자 내부에 실제 검색량 점수를 표시하고, YlOrRd(노랑-주황-빨강) 색상 흐름을 적용합니다.
sns.heatmap(df_2025_matrix, annot=True, fmt="d", cmap="YlOrRd", linewidths=0.5, cbar_kws={'label': '검색 관심도 지표'})

plt.title("2025년 전국 16개 시도별 계절가전(냉방 vs 난방) 검색 관심도 패턴 히트맵 분석", fontsize=14, pad=15)
plt.xlabel("계절 가전 품목", fontsize=12)
plt.ylabel("행정구역 (지역)", fontsize=12)
plt.tight_layout()
plt.show()
```

```python
target_data = df_trends[['에어컨_2025', '선풍기_2025', '전기장판_2025']]
target_data.columns = ['에어컨 (2025)', '선풍기 (2025)', '전기장판 (2025)']

# Seaborn을 활용한 박스 플롯 그리기
sns.boxplot(data=target_data, palette="Set2", width=0.5)
sns.stripplot(data=target_data, color="black", size=5, alpha=0.4, jitter=0.1) # 각 지역별 실제 데이터 포인트를 겹쳐서 표시

plt.title("2025년 주요 계절가전별 전국 검색량 분포 및 편차 비교", fontsize=14, pad=15)
plt.ylabel("검색량 지표 (Relative Search Volume)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.6)
plt.tight_layout()
plt.show()
```

```python
df_trends_sorted = df_trends.sort_values(by='에어컨_2025', ascending=True)

# 그래프 시각화
fig, ax = plt.subplots(figsize=(12, 8))

y_indices = range(len(df_trends_sorted))
bar_width = 0.35

ax.barh([y - bar_width/2 for y in y_indices], df_trends_sorted['에어컨_2025'], bar_width, label='에어컨 (2025)', color='#ff4d6d')
ax.barh([y + bar_width/2 for y in y_indices], df_trends_sorted['전기장판_2025'], bar_width, label='전기장판 (2025)', color='#0077b6')

ax.set_yticks(y_indices)
ax.set_yticklabels(df_trends_sorted['geoName'], fontsize=11)
ax.set_xlabel('상대적 검색 관심도 (0-100)', fontsize=12)
ax.set_title('대한민국 17개 시도별 에어컨 vs 전기장판 검색 관심도 (2025년)', fontsize=16, fontweight='bold', pad=15)
ax.legend(fontsize=12)
ax.grid(axis='x', linestyle='--', alpha=0.6)

plt.tight_layout()
plt.show()
```
