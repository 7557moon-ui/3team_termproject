```python
#계절가전 하위 카테고리의 상품별 분기 별 클릭 수 평균 그래프를 그린다.
#season_df를 기반으로 category3과 quarter(분기)로 groupby해서 평균 클릭 수(quarterly_click_mean)를 구하고
#quarterly_click_mean를 기반으로 그래프를 분기 별 추이를 그린다.

import matplotlib.pyplot as plt

quarter_order = ["Q1", "Q2", "Q3", "Q4"]

quarterly_click_mean = (
    season_df
    .groupby(["category3", "quarter"])["quarterly_avg_click_ratio"]
    .mean()
    .unstack("category3")
    .reindex(quarter_order)
)

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False

ax = quarterly_click_mean.plot(
    figsize=(15, 8),
    marker="o",
    linewidth=2
)

ax.set_title("계절가전 상품별 분기 클릭 수 평균", fontsize=16)
ax.set_xlabel("분기")
ax.set_ylabel("평균 클릭 수")
ax.set_xticks(range(len(quarter_order)))
ax.set_xticklabels(quarter_order)
ax.grid(alpha=0.3)

plt.legend(
    title="category3",
    bbox_to_anchor=(1.02, 1),
    loc="upper left"
)

plt.tight_layout()
plt.show()
```

```python
#분기별 특징이 비슷한 1/4분기와 2/3분기 별 클릭 수 평균의 차이를 구하여 상품별 계절 별 상품의 선호 차이를 확인

season_comparison = pd.DataFrame({
    "Q1·Q4 평균": quarterly_click_mean.loc[["Q1", "Q4"]].mean(),
    "Q2·Q3 평균": quarterly_click_mean.loc[["Q2", "Q3"]].mean()
})

season_comparison["클릭 수 차이"] = (
    season_comparison["Q1·Q4 평균"]
    - season_comparison["Q2·Q3 평균"]
)

season_comparison = season_comparison.sort_values("클릭 수 차이")

colors = [
    "red" if value > 0 else "blue"
    for value in season_comparison["클릭 수 차이"]
]

ax = season_comparison["클릭 수 차이"].plot(
    kind="barh",
    figsize=(11, 9),
    color=colors
)

ax.axvline(0, color="black", linewidth=1)
ax.set_title("상품별 Q1·Q4와 Q2·Q3 클릭 수 평균 차이", fontsize=16)
ax.set_xlabel("Q1·Q4 평균 - Q2·Q3 평균")
ax.set_ylabel("상품")
ax.grid(axis="x", alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

click_col = "quarterly_avg_click_ratio"
quarter_order = ["Q1", "Q2", "Q3", "Q4"]

# 결측값과 비정상 클릭량 지수 100 제외
valid_season_df = season_df[
    season_df[click_col].notna()
    & season_df[click_col].ne(100)
].copy()

# 상품별 분기 평균 클릭량
season_product_clicks = (
    valid_season_df
    .groupby(["category3", "quarter"])[click_col]
    .mean()
    .unstack("quarter")
    .reindex(columns=quarter_order)
    .dropna()
)

# 그래프에 사용할 지표 생성
season_product_clicks["겨울_평균"] = (
    season_product_clicks[["Q1", "Q4"]].mean(axis=1)
)

season_product_clicks["여름_평균"] = (
    season_product_clicks[["Q2", "Q3"]].mean(axis=1)
)

season_product_clicks["분기별_차이"] = (
    season_product_clicks[quarter_order].max(axis=1)
    - season_product_clicks[quarter_order].min(axis=1)
)

season_product_clicks["최고_분기"] = (
    season_product_clicks[quarter_order].idxmax(axis=1)
)

season_product_clicks = season_product_clicks.reset_index()
```

```python
#다변량 패턴 군집도 시각화
#히트맵을 사용하여, 어떤 상품이 특정 계절에 클릭량이 군집되어있는지 확인 가능
#대부분의 계절 가전의 경우, 2/3분기와 1/4분기에 비슷한 군집도를 형성
#공기청정기와 업소용 냉온풍기의 경우에만 다른 패턴을 보임

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


valid_season_df = season_df[
    season_df["quarterly_avg_click_ratio"].notna()
    & season_df["quarterly_avg_click_ratio"].ne(100)
].copy()

season_click_pivot = (
    valid_season_df
    .groupby(["category3", "quarter"])["quarterly_avg_click_ratio"]
    .mean()
    .unstack("quarter")
    .reindex(columns=["Q1", "Q2", "Q3", "Q4"])
    .dropna()
)

# 상품별 평균과 표준편차를 이용한 표준화
season_click_standardized = season_click_pivot.sub(
    season_click_pivot.mean(axis=1),
    axis=0
).div(
    season_click_pivot.std(axis=1).replace(0, 1),
    axis=0
)

sns.clustermap(
    season_click_standardized,
    cmap="RdBu_r",
    center=0,
    figsize=(10, 14),
    linewidths=0.3,
    col_cluster=False,
    standard_scale=None
)

plt.suptitle(
    "계절가전 상품별 분기 클릭량 패턴 군집도",
    y=1.02,
    fontsize=16
)

plt.show()
```

```python
#위에서 보듯이 대부분의 의류의 경우에는 2,3분기의 관심도가 1,4분기의 관심도보다 높지만, 정작 판매 총액은 4분기에 매우 높아짐
#클릭 수는 판매 수가 아니며 클릭 수 지표는 기본적으로 한 해의 가장 높았던 클릭 수를 100으로 잡은 상대 지표이기 때문에 정확한 판매 추이 지표라고 보기는 힘듬
#또한 예상 인사이트에서 1,4분기에 높은 판매량을 기록한다고 예측했던 점퍼나 아우터같은 의류는 판매단가가 다른 의류에 비해 높다는 점도 영향을 끼쳤다고 예상

clothing_df = df[df["category1"] == "패션의류"]

quarter_order = ["Q1", "Q2", "Q3", "Q4"]

quarterly_sales = (
    clothing_df
    .groupby("quarter")["sales_amount"]
    .first()
    .reindex(quarter_order)
)

quarterly_sales.reset_index(name="sales_amount")
```

```python
#추가 인사이트 도출을 위해 데이터프레임을 추가 분석
#분기간 클릭량 평균의 차이가 가장 큰 상품을 도출한다.
import numpy as np

quarter_order = ["Q1", "Q2", "Q3", "Q4"]
click_col = "quarterly_avg_click_ratio"

# 상품별·분기별 평균 클릭량 계산
quarterly_product_clicks = (
    df.groupby(["category1", "category2", "category3", "quarter"])[click_col]
    .mean()
    .unstack("quarter")
    .reindex(columns=quarter_order)
)

# 분기 간 차이 지표 계산
quarterly_product_clicks["highest_quarter"] = (
    quarterly_product_clicks[quarter_order].idxmax(axis=1)
)

quarterly_product_clicks["lowest_quarter"] = (
    quarterly_product_clicks[quarter_order].idxmin(axis=1)
)

quarterly_product_clicks["max_click"] = (
    quarterly_product_clicks[quarter_order].max(axis=1)
)

quarterly_product_clicks["min_click"] = (
    quarterly_product_clicks[quarter_order].min(axis=1)
)

quarterly_product_clicks["click_difference"] = (
    quarterly_product_clicks["max_click"]
    - quarterly_product_clicks["min_click"]
)

# 상품 평균 대비 변동 정도
quarterly_product_clicks["variation_ratio"] = (
    quarterly_product_clicks["click_difference"]
    / quarterly_product_clicks[quarter_order].mean(axis=1).replace(0, np.nan)
)

# 분기 데이터가 모두 존재하는 상품만 선택
large_difference_products = (
    quarterly_product_clicks
    .dropna(subset=quarter_order)
    .sort_values("click_difference", ascending=False)
    .reset_index()
)


large_difference_products.sort_values(
    "variation_ratio",
    ascending=False
).head(20)
```

```python
#위의 데이터로 어떤 상품이 어떤 분기에 가장 많은 관심을 끌고, 또 그럴 수 밖에 없는 이유를 설명할 수 있다.
#목도리, 귀마개의 경우는 평균 기온이 매우 떨어지는 4분기와 1분기를 제외하고는 사용하기 힘든 용품에 해당하며,
#금꽃은 2분기가 제철이며, 절임배추는 흔히 말하는 김장철과 겹치는 4분기를 제외하고는 많이 사용되지 않는다.
#또한 특정 계절에 선호되는 물품 중에서도, 분기별로 큰 차이가 나는 상품도 존재한다.

#위 데이터프레임에서 확인할 수 있듯 스키 관련 물품들이 많은데, 1,4분기와 2,3분기의 차이가 매우 크지만 1분기가 4분기에 비해서 매우 높다.
#이는 스키는 기온이 매우 낮아지는 12월~2월 사이에 집중되는 스포츠이기 때문에, 기온이 상대적으로 높은 10~11월에는 거의 수요가 없을 것이다.
#스키장은 11월 말~12월 중순쯤 개방하는 경우가 많고, 일반적으로 평균 기온이 가장 낮은 1월에 수요가 가장 많을 것
```

```python
#기온과 스키 물품들의 상관관계를 분석하기 위해, 월별 클릭 지표를 새로운 데이터프레임으로 해서 분석을 시도한다.
#월별 클릭 지표 파일을 분기별 클릭 지표 데이터프레임과 동일한 형태의 컬럼으로 만든다.
monthly_df = pd.read_csv(
    "../data/processed/naver_shopping_3depth_monthly_2025.csv"
)

monthly_df = monthly_df.rename(columns={
    "parent1_name": "category1",
    "parent2_name": "category2",
    "category_name": "category3",
    "parent1_code": "category1_id",
    "parent2_code": "category2_id",
    "category_code": "category3_id",
    "click_ratio": "monthly_click_ratio"
})

monthly_df["period"] = pd.to_datetime(monthly_df["period"])
monthly_df["month"] = pd.to_numeric(
    monthly_df["month"],
    errors="coerce"
)

# 결측값과 중복 행 제거
monthly_df = (
    monthly_df
    .dropna(subset=[
        "category1",
        "category2",
        "category3",
        "month",
        "monthly_click_ratio"
    ])
    .drop_duplicates()
    .reset_index(drop=True)
)

monthly_df.head()
```

```python
#기상청이 제공한 2025년의 월별 평균 기온 자료를 가져와 데이터프레임화한다.
temperature_df = pd.read_csv(
    "../data/raw/temperature_seoul_monthly_2025.csv",
    encoding="cp949",
    skiprows=3
)

temperature_df["년월"] = (
    temperature_df["년월"]
    .astype(str)
    .str.extract(r"(\d{4}-\d{2})")[0]
)

temperature_df["년월"] = pd.to_datetime(
    temperature_df["년월"],
    errors="coerce"
)

temperature_df["평균기온(℃)"] = pd.to_numeric(
    temperature_df["평균기온(℃)"],
    errors="coerce"
)

temperature_df = temperature_df.dropna(
    subset=["년월", "평균기온(℃)"]
)

temperature_df["year"] = temperature_df["년월"].dt.year
temperature_df["month"] = temperature_df["년월"].dt.month

monthly_temperature = temperature_df[
    ["year", "month", "평균기온(℃)"]
].reset_index(drop=True)

monthly_temperature
```

```python
#스키상품의 평균클릭량을 계산한다.
ski_monthly_df = monthly_df[
    monthly_df["category2"] == "스키/보드"
].copy()

monthly_ski_clicks = (
    ski_monthly_df
    .groupby("month", as_index=False)["monthly_click_ratio"]
    .mean()
    .rename(columns={
        "monthly_click_ratio": "스키상품_평균클릭량"
    })
)

monthly_ski_clicks  
```

```python
#스키상품의 모든 종류를 검색
ski_monthly_df["category3"].unique()
```

```python
#스키상품 평균클릭량과 평균기온 데이터프레임을 merge해서 합병하고, 월별로 정렬한다.
ski_temperature_analysis = pd.merge(
    monthly_ski_clicks,
    monthly_temperature,
    on="month",
    how="inner"
).sort_values("month")

ski_temperature_analysis
```

```python
from scipy.stats import pearsonr

correlation = ski_temperature_analysis[
    ["평균기온(℃)", "스키상품_평균클릭량"]
].corr().loc[
    "평균기온(℃)",
    "스키상품_평균클릭량"
]

# p-value 계산
_, p_value = pearsonr(
    ski_temperature_analysis["평균기온(℃)"],
    ski_temperature_analysis["스키상품_평균클릭량"]
)

print(f"상관계수: {correlation:.3f}")
print(f"p-value: {p_value:.3f}")
#상관계수와 p-value의 출력 결과, 강한 음의 관계를 가지고 유의미한 상관관계를 가진다.
```
