# 온라인 쇼핑몰 판매매체별 거래액

```python
# 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 원본 데이터 로드
df_sales = pd.read_csv("../data/raw/online_sales_by_channel_2025.csv", encoding='utf-8')

# 컬럼명 직관적 변경
rename_columns_1 = {
    '상품군별(1)': '상품군_대분류',
    '상품군별(2)': '상품군_소분류',
    '판매매체별(1)': '판매매체',
    '2026.1/4 p)': '2026.1/4'  # 임시치 표시 제거
}

# 정의한 사전을 데이터프레임에 적용
df_sales = df_sales.rename(columns=rename_columns_1)

# 전처리 완료된 정제 파일을 깨짐 방지 포맷으로 저장
df_sales.to_csv("cleaned_online_sales.csv", index=False, encoding='utf-8-sig')

df_sales
```

```python
# 예상 인사이트: 가전·전자·통신기기의 경우 PC 중심의 인터넷 쇼핑보다 모바일 쇼핑의 거래액 규모가 훨씬 클 것이다.
# 해당 가설을 검증하기 위해 대분류 카테고리가 '가전·전자·통신기기'이고 소계 데이터인 행만 필터링하여 데이터프레임을 만든다.
appliance_df = df_sales[(df_sales["상품군_대분류"] == "가전·전자·통신기기") & (df_sales["상품군_소분류"] == "소계")]

appliance_df
```

```python
# 한글 폰트 깨짐 방지 및 마이너스 기호 설정 (Windows 환경 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# '계' 행을 제외하고 '인터넷쇼핑'과 '모바일쇼핑' 행만 선택
media_plot_df = appliance_df[appliance_df["판매매체"].isin(["인터넷쇼핑", "모바일쇼핑"])]

# 분기 컬럼들만 리스트로 추출
quarter_cols = [col for col in media_plot_df.columns if '/' in col]

plt.figure(figsize=(13, 6))

# 각 매체별로 선 그래프 시각화
for idx, row in media_plot_df.iterrows():
    plt.plot(quarter_cols, row[quarter_cols], marker='o', label=row["판매매체"], linewidth=2.5)

plt.title("가전·전자·통신기기 상품군의 판매매체별(인터넷 vs 모바일) 거래액 추이", fontsize=14, pad=15)
plt.xlabel("분기", fontsize=12)
plt.ylabel("거래액 (백만 원)", fontsize=12)
plt.xticks(rotation=45)
plt.legend(title="판매매체", fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
clothing_total = df_sales[(df_sales["상품군_대분류"] == "의복") & (df_sales["상품군_소분류"] == "소계") & (df_sales["판매매체"] == "계")]
appliance_total = df_sales[(df_sales["상품군_대분류"] == "가전·전자·통신기기") & (df_sales["상품군_소분류"] == "소계") & (df_sales["판매매체"] == "계")]

# 분기 컬럼명 리스트 추출
quarter_cols = [col for col in df_sales.columns if '/' in col]

# 3. 시계열 추이 선 그래프 시각화
plt.figure(figsize=(14, 6))

plt.plot(quarter_cols, clothing_total[quarter_cols].iloc[0], marker='o', color='blue', linewidth=2.5, label='의복 (의류)')
plt.plot(quarter_cols, appliance_total[quarter_cols].iloc[0], marker='s', color='orange', linewidth=2.5, label='가전·전자·통신기기 (난방/냉방기구)')

plt.title("의류 및 가전 품목별 분기별 총 거래액 추이 비교 (2022~2026)", fontsize=14, pad=15)
plt.xlabel("분기", fontsize=12)
plt.ylabel("총 거래액 (백만 원)", fontsize=12)
plt.xticks(rotation=45)
plt.grid(True, linestyle='--', alpha=0.5)
plt.legend(fontsize=11)
plt.tight_layout()
plt.show()

# 의복과 가전 품목의 장기 거래액 추이를 비교해 보면 두 상품군 모두 매년 4분기(겨울철 기틀)에 거래액이 정점을 찍는 강력한 계절성 동기화 패턴을 보인다.
# 의류의 경우 4분기 헤비 아우터 집중 구매가 원인이며, 가전 제품군 역시 4분기 블랙프라이데이 연말 프로모션과 고단가 동절기 난방 기기 수요가 맞물려 실제 거래 금액의 피크를 형성함을 정량적으로 입증할 수 있다.
```
