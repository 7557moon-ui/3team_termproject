# 지역별 상품군별 온라인쇼핑 해외직접판매액 정리

```python
# 라이브러리
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
```

```python
# 원본 데이터 로드
df_export = pd.read_csv("../data/raw/overseas_direct_sales_by_region.csv", encoding='utf-8')

# 컬럼명 직관적 변경
df_export = df_export.rename(columns={
    '지역별(1)': 'continent',
    '지역별(2)': 'country',
    '지역별(3)': 'region_detail',
    '상품군별(1)': 'category1',
    '상품군별(2)': 'category2',
    '판매유형별(1)': 'export_type',
    '2026.1/4 p)': '2026.1/4'
})

df_export
```

```python
# info()를 사용하여 데이터 타입과 결측치(NaN) 존재 여부를 확인한다.
# 2022년도 초기 데이터 중 일부 분기에 결측치가 존재하는 것을 파악할 수 있다.
df_export.info()
```

```python
# NaN값(결측값)을 제외하고 분석을 진행하기 위해 결측치를 0으로 대치하는 정제 과정을 진행한다.
# 불필요한 공백도 제거

quarter_cols = [col for col in df_export.columns if '/' in col]
df_export[quarter_cols] = df_export[quarter_cols].fillna(0)

# 문자열 공백 전처리
text_cols = ['continent', 'country', 'category1', 'category2', 'export_type']
for col in text_cols:
    df_export[col] = df_export[col].str.replace(" ", "").str.strip()

df_export
# 결측치 정제 완료 후 안정적인 데이터프레임 확보 완료.
```

```python
# 예상 인사이트: 아시아 국가(중국, 일본)와 북미(미국) 시장 간의 주요 역직구 구매 상품군(category1) 비중은 뚜렷하게 다를 것이다.
# 가설 검증을 위해 대륙 및 국가의 합계 행을 제외한 실제 국가별 상품군 판매 데이터프레임을 생성한다.
valid_export_df = df_export[
    (df_export['continent'] != '합계') & 
    (df_export['country'] != '소계') & 
    (df_export['category1'] != '합계') & 
    (df_export['export_type'] == '계')
]

valid_export_df
```

```python
# 한글 폰트 깨짐 방지 및 마이너스 기호 설정 (Windows 환경 기준)
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 상품군별(1) 항목이 '화장품'인 데이터만 추출
clothing_export_df = valid_export_df[valid_export_df["category1"] == "의류및패션관련상품"]

# 국가별(지역별(2))로 2025년 4분기 거래액을 그룹화하여 합산한 뒤 상위 5개국 정렬
top_clothing_countries = clothing_export_df.groupby("country")["2025.4/4"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
top_clothing_countries.plot(kind='bar', color='lightblue', edgecolor='black', width=0.5)

plt.title("2025년 4분기 주요 국가/지역별 의류 및 패션 관련 상품 해외 직접 판매액 (역직구)", fontsize=14, pad=15)
plt.xlabel("국가 및 지역", fontsize=12)
plt.ylabel("해외직접판매액 (백만 원)", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# 의류 및 패션 관련 상품의 역직구 거래액을 분석한 결과, 아시아 권역 및 특정 주요 국가로의 수출 집중도가 높게 나타난다.
# 글로벌 K-패션 플랫폼의 확장에 따라 국가별 선호 의류 스타일과 계절적 요인을 고려한 역직구 타겟팅 전략이 유효함을 시사한다.
```

```python
# 원본 품목명인 '가전·전자·통신기기' 데이터만 추출
appliance_export_df = valid_export_df[valid_export_df["category1"] == "가전·전자·통신기기"]

# 국가별(지역별(2))로 2025년 4분기(겨울철 난방도구 집중기) 거래액을 그룹화하여 합산한 뒤 상위 5개국 정렬
top_appliance_countries = appliance_export_df.groupby("country")["2025.4/4"].sum().sort_values(ascending=False).head(5)

plt.figure(figsize=(10, 6))
top_appliance_countries.plot(kind='bar', color='thistle', edgecolor='black', width=0.5)

plt.title("2025년 4분기 주요 국가/지역별 가전·전자·통신기기 해외 직접 판매액 (역직구)", fontsize=14, pad=15)
plt.xlabel("국가 및 지역", fontsize=12)
plt.ylabel("해외직접판매액 (백만 원)", fontsize=12)
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
# 공백 제거 및 주요 거점 국가 필터링
df_export['category1'] = df_export['category1'].str.replace(" ", "")
df_export['country'] = df_export['country'].str.replace(" ", "")
df_export['export_type'] = df_export['export_type'].str.replace(" ", "")

target_countries = ['미국', '중국', '일본', '아세안']
fashion_df = df_export[
    (df_export['category1'] == '의류및패션관련상품') & 
    (df_export['country'].isin(target_countries)) & 
    (df_export['export_type'] == '계')
]

# 2025년 각 분기 컬럼 선택 및 피벗 테이블 생성
quarters_2025 = ['2025.1/4', '2025.2/4', '2025.3/4', '2025.4/4']
pivot_fashion = fashion_df.pivot_table(index='country', values=quarters_2025, aggfunc='sum')
pivot_fashion = pivot_fashion[quarters_2025] # 분기 순서 정렬

# 100% 누적 플롯을 위한 행별 비율 변환
pivot_fashion_ratio = pivot_fashion.div(pivot_fashion.sum(axis=1), axis=0) * 100

# 시각화
ax = pivot_fashion_ratio.plot(kind='bar', stacked=True, figsize=(11, 6), cmap='Set3', edgecolor='black', width=0.4)
plt.title("주요 거점 국가별 K-패션 역직구 매출액의 분기별 비중 대조 (2025)", fontsize=14, pad=15)
plt.xlabel("해외 거점 국가/지역", fontsize=12)
plt.ylabel("분기별 매출 비중 (%)", fontsize=12)
plt.legend(['1분기', '2분기', '3분기', '4분기'], bbox_to_anchor=(1.02, 1), loc='upper left')
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
