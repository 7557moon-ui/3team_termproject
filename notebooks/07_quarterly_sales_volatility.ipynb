```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 및 마이너스 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 로드
df_naver = pd.read_csv("../data/processed/final_naver_click_sales_quarterly_2025.csv", encoding='utf-8')

# 분석 대상 주요 소비재 대분류(sales_group) 필터링
target_groups = ['가전·전자·통신기기', '의복', '화장품', '음·식료품', '가구']
df_filtered = df_naver[df_naver['sales_group'].isin(target_groups)]

# 바이올린 플롯 시각화
plt.figure(figsize=(12, 6))
sns.violinplot(
    x='sales_group', 
    y='sales_amount', 
    hue='sales_group',
    legend=False,
    data=df_filtered, 
    palette='muted', 
    inner='quartile', 
    linewidth=1.5
)

plt.title("주요 이커머스 품목별 분기 판매 금액의 통계적 분포 및 변동성 비교", fontsize=14, pad=15)
plt.xlabel("상품군 분류", fontsize=12)
plt.ylabel("분기별 판매 금액 (sales_amount)", fontsize=12)
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# [결과해석 주석셀]
# 바이올린 플롯의 형태가 세로로 길고 팽창해 있을수록 분기별 기후 변화나 계절성 요인에 의해 실제 매출액의 변동 폭과 리스크가 매우 크다는 것을 뜻한다.
# 반면 모양이 납작하고 특정 구간에 밀집해 있는 상품군은 연간 공급과 수요가 매우 안정적으로 유지되는 품목임을 보여준다.
# 이 분석 결과는 공급망 레이아웃 설계 시 재고 확보 가중치를 유동적으로 조절해야 하는 '고위험 기후 민감 품목(예: 패션의류, 가전)'과 상시 안정 품목을 구분 짓는 명확한 통계적 근거가 된다.
```
