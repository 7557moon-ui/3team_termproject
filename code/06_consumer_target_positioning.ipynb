```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 한글 폰트 및 마이너스 깨짐 방지
plt.rcParams['font.family'] = 'Malgun Gothic'
plt.rcParams['axes.unicode_minus'] = False

# 데이터 로드
df_segment = pd.read_csv("../data/processed/naver_shopping_3depth_segment_share_wide_2025.csv", encoding='utf-8')

# 30대와 40대 비율을 합산하여 핵심 소비층(3040_premium) 변수 생성
df_segment['3040_percent'] = df_segment['age_30s_percent'] + df_segment['age_40s_percent']

# 대분류(category1)별로 평균 지표 산출
df_group = df_segment.groupby('category1')[['gender_Female_percent', 'device_Mobile_percent', '3040_percent']].mean().reset_index()

# 버블 차트 시각화
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    df_group['gender_Female_percent'], 
    df_group['device_Mobile_percent'], 
    s=df_group['3040_percent'] * 25,  # 가시성을 위해 버블 크기 가중치 조절
    c=df_group['3040_percent'], 
    cmap='Purples', 
    alpha=0.7, 
    edgecolors='black', 
    linewidths=1.5
)

# 각 버블 옆에 카테고리명 텍스트 라벨 추가
for i, txt in enumerate(df_group['category1']):
    plt.annotate(txt, (df_group['gender_Female_percent'].iloc[i]+0.5, df_group['device_Mobile_percent'].iloc[i]), fontsize=11, weight='bold')

plt.title("이커머스 주요 상품군별 소비 타겟 및 매체 포지셔닝 맵", fontsize=14, pad=15)
plt.xlabel("여성 검색 비율 (%)", fontsize=12)
plt.ylabel("모바일 기기 이용 비율 (%)", fontsize=12)
cbar = plt.colorbar(scatter)
cbar.set_label('3040 연령대 핵심 구매층 비중 (%)', fontsize=11)
plt.grid(True, linestyle='--', alpha=0.5)
plt.tight_layout()
plt.show()

# [결과해석 주석셀]
# 본 시각화 자료는 각 상품군의 성별 관심도와 접속 기기 비중, 그리고 주 소비 연령대를 하나의 평면에 다차원적으로 포지셔닝한 결과이다.
# 상단 우측에 위치할수록 모바일 의존도가 높고 여성 중심적인 성향이 강한 상품군이며, 버블의 크기가 클수록 3040 세대의 구매 가중치가 높음을 의미한다.
# 제안서에서 명시한 '차별화된 홍보 및 이벤트 마케팅' 전개 시, 쿠폰 및 푸시 알림 등의 모바일 최적화 전략을 어떤 카테고리에 우선순위로 매칭해야 하는지 정량적 가이드를 제공한다.
```
