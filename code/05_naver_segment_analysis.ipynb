```python
#데이터프레임의 null 값을 제외한다.
segment_cols = [
    "age_10s_percent",
    "age_20s_percent",
    "age_30s_percent",
    "age_40s_percent",
    "age_50s_percent",
    "age_60s_plus_percent",
    "device_Mobile_percent",
    "device_PC_percent",
    "gender_Female_percent",
    "gender_Male_percent"
]


df = (
    df
    .dropna(subset=segment_cols)
    .reset_index(drop=True)
)
df.isnull().sum()
```

```python
#예상 인사이트: 출산/육아 용품의 경우 3~40대 여성의 비율이 높을 것이다.
#그냥 높은 것에 그치지않고, 다른 카테고리에 비해서도 상대적으로 높은 비율을 가질 것이다.
#해당 인사이트를 도출하기 위해 category1이 출산/육아인 상품의 데이터프레임을 만든다.
baby_df = df[df["category1"] == "출산/육아"]

baby_df
```

```python
age_cols = [
    "age_10s_percent",
    "age_20s_percent",
    "age_30s_percent",
    "age_40s_percent",
    "age_50s_percent",
    "age_60s_plus_percent"
]

age_usage_mean = (
    baby_df[age_cols]
    .mean()
    .reset_index()
)

age_usage_mean.columns = ["연령대", "평균 이용률(%)"]

age_usage_mean
#출산/육아 카테고리는 연령대 별로, 30~40대의 비중이 약 85%에 달할 정도로 매우 높다.
```

```python
gender_cols = [
    "gender_Female_percent",
    "gender_Male_percent"
]

gender_usage_mean = (
    baby_df[gender_cols]
    .mean()
    .rename_axis("성별")
    .reset_index(name="평균 이용률(%)")
)

gender_usage_mean
#성별 역시 여성의 비율이 81%로 남성에 비해 매우 높다.
```

```python
#출산/육아 카테고리와 30/40대, 여성 이용률과의 상관관계를 분석하기 위해 새로운 데이터프레임 생성
#다른 상품군들과의 비교를 위해서 출산/육아 카테고리와 구분하는 컬럼을 추가

analysis_cols = [
    "category1",
    "age_30s_percent",
    "age_40s_percent",
    "gender_Female_percent"
]

category_correlation_df = (
    df[analysis_cols]
    .dropna()
    .copy()
)

# 출산/육아 상품이면 1, 다른 상품이면 0
category_correlation_df["is_baby_category"] = (
    category_correlation_df["category1"] == "출산/육아"
).astype(int)

# 30대와 40대 비율 합계
category_correlation_df["age_30_40_percent"] = (
    category_correlation_df["age_30s_percent"]
    + category_correlation_df["age_40s_percent"]
)

category_correlation_df.head()
```

```python
correlation_results = category_correlation_df[
    [
        "is_baby_category",
        "age_30s_percent",
        "age_40s_percent",
        "age_30_40_percent",
        "gender_Female_percent"
    ]
].corr()["is_baby_category"].drop("is_baby_category")

correlation_results
#상관계수 결과, 30대와는 강한 양의 상관관계, 여성과는 상대적으로 약하지만 유의미한 양의 상관관계를 가진다.
#다만 40대와는 매우 약한 음의 상관관계를 지니며, 상관계수 상으로는 40대와의 연결고리가 매우 약하다.
```

```python
from scipy.stats import pearsonr
import pandas as pd

p_values = {}

for column in correlation_results.index:
    _, p_value = pearsonr(
        category_correlation_df["is_baby_category"],
        category_correlation_df[column]
    )

    p_values[column] = p_value

p_value_results = pd.DataFrame({
    "상관계수": correlation_results,
    "p-value": pd.Series(p_values)
})

p_value_results["통계적 유의성"] = (
    p_value_results["p-value"] < 0.05
)

p_value_results
#p-value가 모두 0.05 이하기 때문에 통계적으로 유의미한 상관계수를 가진다.
#따라서 출산/육아 카테고리는 30대, 여성 이용률과 유의미한 상관관계를 가지지만, 40대 이용률과는 큰 상관관계를 가지지 못 한다.
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.rcParams["font.family"] = "Malgun Gothic"
plt.rcParams["axes.unicode_minus"] = False


correlation_plot_df = (
    p_value_results
    .reset_index()
    .rename(columns={"index": "이용자 특성"})
)

correlation_plot_df["이용자 특성"] = correlation_plot_df["이용자 특성"].replace({
    "age_30s_percent": "30대",
    "age_40s_percent": "40대",
    "age_30_40_percent": "30·40대",
    "gender_Female_percent": "여성"
})

# 상관관계 방향에 따른 색상 지정
colors = [
    "tomato" if value > 0 else "steelblue"
    for value in correlation_plot_df["상관계수"]
]

plt.figure(figsize=(10, 6))

bars = plt.bar(
    correlation_plot_df["이용자 특성"],
    correlation_plot_df["상관계수"],
    color=colors,
    alpha=0.8
)


for bar, correlation, p_value in zip(
    bars,
    correlation_plot_df["상관계수"],
    correlation_plot_df["p-value"]
):

    y_position = correlation + 0.03 if correlation >= 0 else correlation - 0.06



plt.axhline(0, color="black", linewidth=1)
plt.ylim(-0.2, 1.0)

plt.title("출산·육아 카테고리와 이용자 특성의 상관관계", fontsize=15)
plt.xlabel("이용자 특성")
plt.ylabel("상관계수")
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
comparison_plot_df = category_correlation_df.copy()

comparison_plot_df["상품군"] = comparison_plot_df["is_baby_category"].map({
    0: "기타 상품군",
    1: "출산·육아"
})

comparison_plot_df = comparison_plot_df.melt(
    id_vars="상품군",
    value_vars=[
        "age_30s_percent",
        "age_40s_percent",
        "gender_Female_percent"
    ],
    var_name="이용자 특성",
    value_name="이용률"
)

comparison_plot_df["이용자 특성"] = comparison_plot_df["이용자 특성"].replace({
    "age_30s_percent": "30대",
    "age_40s_percent": "40대",
    "gender_Female_percent": "여성"
})

plt.figure(figsize=(12, 7))

sns.boxplot(
    data=comparison_plot_df,
    x="이용자 특성",
    y="이용률",
    hue="상품군",
    palette={
        "기타 상품군": "lightgray",
        "출산·육아": "tomato"
    }
)

plt.title("출산·육아와 기타 상품군의 이용자 특성 비교", fontsize=15)
plt.xlabel("이용자 특성")
plt.ylabel("이용률(%)")
plt.grid(axis="y", alpha=0.3)
plt.legend(title="상품군")
plt.tight_layout()
plt.show()
#박스플롯의 결과로 보았을 때, 출산/육아 카테고리의 30대와 여성 이용률은 다른 카테고리와 비교했을 때 높은 중앙값을 가진다.(다른 상품군의 일부 데이터 역시 높은 값을 가짐)
#하지만, 40대 이용률의 경우 두 중앙값이 비슷해 다른 카테고리와 크게 차이나지 않는다.
#따라서, 출산/육아 상품의 경우 30대와 여성의 이용률이 다른 카테고리에 비해 전반적으로 높지만, 40대는 다른 카테고리와 비슷한 비중을 가진다.
```
